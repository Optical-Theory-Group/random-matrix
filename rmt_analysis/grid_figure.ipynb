```python
x_min = -1.05
x_max = 1.05
y_min = -1.05
y_max = 1.05
```

```python
# -----------------------------------------------------------------------------
# Polar grids
# -----------------------------------------------------------------------------

# Irregular polar grid from random r and t values
dr = 0.1
dt_divisor = 20
mg = mode_grid_factory.from_dr_dt(
    dr, 2 * np.pi / dt_divisor, include_central_mode=False, r_lim=1.0, rotation_angle=2 * np.pi / (2*dt_divisor)
)
fig, ax = mg.plot()
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
ax.plot([-1, 1], [0, 0], linestyle='--', linewidth=1, zorder=0,color="black")
ax.plot([0, 0], [-1, 1], linestyle='--', linewidth=1, zorder=0,color="black")
fig.savefig("/home/nbyrnes/code/random-matrix/rmt_analysis/grid_polar.svg",format="svg")
```

```python
# -----------------------------------------------------------------------------
# Tilings
# -----------------------------------------------------------------------------

# Hexagons
mg = mode_grid_factory.from_tiling(
    tiling_type="hexagons",
    side_length=0.1,
    r_lim=1.0,
    grid_wave_type="all",
    rotation_angle=0.0,
    translation_vector=np.array([0.0, 0.0]),
)
fig, ax = mg.plot()
ax.set_xlim(x_min, x_max)
ax.set_ylim(y_min, y_max)
ax.plot([-1, 1], [0, 0], linestyle='--', linewidth=1, zorder=0,color="black")
ax.plot([0, 0], [-1, 1], linestyle='--', linewidth=1, zorder=0,color="black")
fig.savefig("/home/nbyrnes/code/random-matrix/rmt_analysis/grid_hexagon.svg",format="svg")

```
